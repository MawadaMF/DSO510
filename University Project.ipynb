```python
import pandas as pd
import seaborn as sns
acceptance_rate=pd.read_csv('acceptance.csv')
avg_salary=pd.read_csv('salary.csv')
```

```python
acceptance_rate.head(15)
```

```python
avg_salary.head(15)
```

```python
acceptance_rate['business_school']=acceptance_rate['School'].map\
                                   (lambda school: school[school.index('(')+1:-1] if '(' in school else school)
```

```python
acceptance_rate.head(15)
```

```python
avg_salary['business_school']=avg_salary['School'].map\
                                   (lambda school: school[school.index('(')+1:-1] if '(' in school else school)
```

```python
avg_salary.head(15)
```

```python
print(acceptance_rate.shape)
print(avg_salary.shape)
```

```python
df=acceptance_rate.merge(avg_salary, how='inner', on='business_school')
df.shape
```

```python
df.head(10)
```

```python
unmatched_acceptance = acceptance_rate[acceptance_rate['business_school'].isin(df['business_school']) == False]
len(unmatched_acceptance)
```

```python
unmatched_salary = avg_salary[avg_salary['business_school'].isin(df['business_school']) == False]
len(unmatched_salary)
```

```python
df['Acceptance Rate'] = df['Acceptance Rate'].map(lambda rate: float(rate[:rate.index('%')])/100)
```

```python
df['Average Salary'] = df['Average Salary'].map(lambda salary: int(salary[1:salary.index(',')]+
                                                                   salary[salary.index(',')+1:salary.index(' ')]))
```

```python
df.head(10)
```

```python
df['Acceptance Rate'].corr(df['Average Salary'])
```

```python
sns.scatterplot(x='Acceptance Rate', y='Average Salary', data=df)
```
