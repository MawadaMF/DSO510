```python
import pandas as pd
import seaborn as sns
acceptance_rate=pd.read_csv('acceptance.csv')
avg_salary=pd.read_csv('salary.csv')
```

```python
acceptance_rate.head(15)
```

```python
avg_salary.head(15)
```

```python
acceptance_rate['business_school']=acceptance_rate['School'].map\
                                   (lambda school: school[school.index('(')+1:-1] if '(' in school else school)
```

```python
acceptance_rate.head(15)
```

```python
avg_salary['business_school']=avg_salary['School'].map\
                                   (lambda school: school[school.index('(')+1:-1] if '(' in school else school)
```

```python
avg_salary.head(15)
```

```python
print(acceptance_rate.shape)
print(avg_salary.shape)
```

```python
df=acceptance_rate.merge(avg_salary, how='inner', on='business_school')
df.shape
```

```python
df.head(10)
```

```python
unmatched_acceptance=acceptance_rate[acceptance_rate['business_school'].isin(df['business_school'])==False]
len(unmatched_acceptance)
```

```python
unmatched_salary=avg_salary[avg_salary['business_school'].isin(df['business_school'])==False]
len(unmatched_salary)
```

```python
df['Acceptance Rate']=df['Acceptance Rate'].map(lambda rate: float(rate[:rate.index('%')])/100)
```

```python
df['Average Salary']=df['Average Salary'].map(lambda salary: int(salary[1:salary.index(',')]+\
                                              salary[salary.index(',')+1:salary.index(' ')]))
```

```python
df.head(10)
```

```python
df['Acceptance Rate'].corr(df['Average Salary'])
```

```python
sns.scatterplot(x='Acceptance Rate',y='Average Salary',data=df)
```

```python
unmatched_salary[['School','business_school']]
```

```python
drop=['University','of','State','College','Business','School','The']
unmatched_acceptance['business_school']=unmatched_acceptance['business_school'].map(lambda school: school.split(' '))
for word in drop:
    unmatched_acceptance['business_school'].map(lambda school: school.remove(word) if word in school else school)
unmatched_acceptance['business_school']=unmatched_acceptance['business_school'].map(lambda school: ' '.join(school))
```

```python
unmatched_acceptance[['School','business_school']]
```

```python
unmatched_acceptance[['School','business_school']]
```

```python
cleaned=unmatched_acceptance.merge(unmatched_salary,how='inner',on='business_school')
cleaned
```

```python
still_unmatched_a=unmatched_acceptance[unmatched_acceptance['business_school'].isin(cleaned['business_school'])==False]
still_unmatched_s=unmatched_salary[unmatched_salary['business_school'].isin(cleaned['business_school'])==False]
```

```python
still_unmatched_a[['School','business_school']]
```

```python
still_unmatched_s[['School','business_school']]
```

```python
still_unmatched_a.loc[[53,84],'business_school']=['TCNJ','Donahue']
still_unmatched_s.loc[[29,31,32,58,60,64,73],'business_school']=['Miami','Darla Moore','Chaifetz','Marriott','Lerner',
                                                                 'Jack Welch','Trulaske']
```

```python
matched=still_unmatched_a.merge(still_unmatched_s,how='inner',on='business_school')
matched
```

```python
cleaned['Acceptance Rate']=cleaned['Acceptance Rate'].map(lambda rate: float(rate[:rate.index('%')])/100)
cleaned['Average Salary']=cleaned['Average Salary'].map(lambda salary: int(salary[1:salary.index(',')]+\
                                              salary[salary.index(',')+1:salary.index(' ')]))
```

```python
matched['Acceptance Rate']=matched['Acceptance Rate'].map(lambda rate: float(rate[:rate.index('%')])/100)
matched['Average Salary']=matched['Average Salary'].map(lambda salary: int(salary[1:salary.index(',')]+\
                                              salary[salary.index(',')+1:salary.index(' ')]))
```

```python
frames=[df,cleaned,matched]
cleaned_df=pd.concat(frames)
```

```python
header=['School_x','business_school','Acceptance Rate','Average Salary']
```
